```python
# Standard scientific Python imports
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import re
from io import StringIO
import seaborn as sns

# Import datasets, classifiers and performance metrics
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_fscore_support
from sklearn.utils.multiclass import unique_labels
from sklearn.metrics import confusion_matrix
from sklearn.datasets import make_hastie_10_2
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import scale
from sklearn.model_selection import KFold

#ROC Curve
from sklearn.svm import SVC
from sklearn.metrics import plot_roc_curve
from sklearn.datasets import make_classification

# scikit-learn k-fold cross-validation
from numpy import array
```

```python
input_data = pd.read_csv("/Users/SailahariMullapudi/Desktop/PanPred-master/test_data/curated_input_GY.csv")
input_data2 = input_data.rename(columns={"Unnamed: 0": "Isolate"})
output_data = pd.read_csv("/Users/SailahariMullapudi/Desktop/PanPred-master/test_data/Metadata.csv")
output_data = output_data.replace(to_replace=['S', 'R'], value=[0, 1])
input_all = input_data2.merge(output_data, on='Isolate', how='left')
del input_all['AMP']
del input_all['CET']
del input_all['TBM']
del input_all['TMP']

input_all.head()
```

# CTZ

```python
input_allna = input_all.dropna()

CTZ = input_allna['CTZ']

del input_allna['CTX']
del input_allna['AMC']
del input_allna['TZP']
del input_allna['CXM']
del input_allna['GEN']
del input_allna['CIP']
del input_allna['AMX_x']
del input_allna['AMX_y']
del input_allna['Isolate']
del input_allna['CTZ']
del input_allna['Year_x']
del input_allna['Year_y']
x = np.array(input_allna)
print(x)

X = x
kf = KFold(n_splits=5)
for train, test in kf.split(X):
    print("%s %s" % (train, test))
```

```python
logreg= LogisticRegression(penalty='l2',solver='lbfgs')
X_train, X_test, CTZ_train, CTZ_test = train_test_split(X, CTZ, test_size=0.2, random_state=42)
logreg.fit(X_train,CTZ_train)
predictedCTZ_LR=logreg.predict(X_test)
predictedCTZ_LR
print(
    f"Classification report for classifier {logreg}:\n"
    f"{metrics.classification_report(CTZ_test, predictedCTZ_LR)}\n"
)

CTZ_cmLR = metrics.ConfusionMatrixDisplay.from_predictions(CTZ_test, predictedCTZ_LR)
CTZ_cmLR.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix:\n{CTZ_cmLR.confusion_matrix}")

plt.show()
```

```python
rfreg = RandomForestClassifier(n_jobs=-1,max_features= "sqrt" ,n_estimators= 100, oob_score = True) 
rf_X_train, rf_X_test, rf_CTZ_train, rf_CTZ_test = train_test_split(X, CTZ, test_size = 0.2, random_state=55)
rfreg.fit(rf_X_train,rf_CTZ_train)
predictedCTZ_RF = rfreg.predict(rf_X_test)
predictedCTZ_RF
print(
    f"Classification report for classifier {rfreg}:\n"
    f"{metrics.classification_report(rf_CTZ_test, predictedCTZ_RF)}\n"
)

CTZ_cmRF = metrics.ConfusionMatrixDisplay.from_predictions(rf_CTZ_test, predictedCTZ_RF)
CTZ_cmRF.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix:\n{CTZ_cmRF.confusion_matrix}")

plt.show()
```

```python
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0)
gbc_X_train, gbc_X_test, gbc_CTZ_train, gbc_CTZ_test = train_test_split(X, CTZ, test_size = 0.2, random_state=55)
gbc.fit(gbc_X_train,gbc_CTZ_train)
predictedCTZ_GBDT = gbc.predict(gbc_X_test)
predictedCTZ_GBDT
print(
    f"Classification report for classifier {gbc}:\n"
    f"{metrics.classification_report(gbc_CTZ_test, predictedCTZ_GBDT)}\n"
)

CTZ_cmGBDT = metrics.ConfusionMatrixDisplay.from_predictions(gbc_CTZ_test, predictedCTZ_GBDT)
CTZ_cmGBDT.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix:\n{CTZ_cmGBDT.confusion_matrix}")

plt.show()
```

```python
clf = svm.SVC(gamma=0.001)
svm_X_train, svm_X_test, svm_CTZ_train, svm_CTZ_test = train_test_split(
    X, CTZ, test_size=0.2, shuffle=True
)
clf.fit(svm_X_train, svm_CTZ_train)
predictedCTZ_SVM = clf.predict(svm_X_test)
predictedCTZ_SVM

print(
    f"Classification report for classifier {clf}:\n"
    f"{metrics.classification_report(svm_CTZ_test, predictedCTZ_SVM)}\n"
)

CTZ_cmSVM = metrics.ConfusionMatrixDisplay.from_predictions(svm_CTZ_test, predictedCTZ_SVM)
CTZ_cmSVM.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix:\n{CTZ_cmSVM.confusion_matrix}")

plt.show()
```

```python
clf_disp = plot_roc_curve(clf, svm_X_test, svm_CTZ_test)
rfreg_disp = plot_roc_curve(rfreg, rf_X_test, rf_CTZ_test, ax=clf_disp.ax_)
logreg_disp = plot_roc_curve(logreg, X_test, CTZ_test, ax=clf_disp.ax_)
gbdt_disp = plot_roc_curve(gbc, gbc_X_test, gbc_CTZ_test, ax=clf_disp.ax_)
plt.title("CTZ")
plt.show()
```
