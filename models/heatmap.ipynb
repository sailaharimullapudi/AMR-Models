```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
output_data = pd.read_csv("/Users/SailahariMullapudi/Desktop/PanPred-master/test_data/Metadata.csv")
R = ['R']
```

```python
CTZ_count = output_data.groupby(['Year', 'CTZ'])['Isolate'].count().reset_index().rename(columns={"CTZ": "Resistance", "Isolate": "CTZ"})
df_sumCTZ = CTZ_count.groupby(['Year']).sum().rename(columns={"CTZ": "SR_CTZ_Sum"})
df_mergeCTZ = pd.merge(CTZ_count, df_sumCTZ,  how='left', on=['Year']).rename(columns={"CTZ": "Antibiotic_Value", "SR_CTZ_Sum": "Antibitoic_Sum"})
rslt_dfCTZ = df_mergeCTZ[df_mergeCTZ['Resistance'].isin(R)]
df_proportionCTZ = rslt_dfCTZ.assign(Proportion = (rslt_dfCTZ.Antibiotic_Value / rslt_dfCTZ.Antibitoic_Sum) * 100).assign(Antibiotic_Name = "CTZ")
df_proportionCTZ
```
